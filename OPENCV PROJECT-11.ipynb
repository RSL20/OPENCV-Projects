```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
def function1():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/smarties.png')
    edge = cv2.Canny(img, 255, 255)
    cv2.imshow('edge', edge)
    cv2.imwrite('/tmp/edge.jpg', edge)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function1()
```

```python
def function2():
    capture = cv2.VideoCapture(0)
    while True:
        ret, frame = capture.read()
        edge = cv2.Canny(frame, 255, 255)
        cv2.imshow('video', edge)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    capture.release()
    cv2.destroyAllWindows()
function2()
```
