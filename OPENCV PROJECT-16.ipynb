```python
def function1():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set2/someshapes.jpg')
    # binarise the image
    edges_detected = cv2.Canny(img, 255, 255)
    contours, hierarchy = cv2.findContours(edges_detected, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    # print(contours)
    for c in contours:
        print(c)
    cv2.imshow('original', img)
    cv2.imshow('edges_detected', edges_detected)
    cv2.imwrite('/tmp/original.jpg', img)
    cv2.imwrite('/tmp/edges_detected.jpg', edges_detected)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function1()
```

```python
def function3():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set2/someshapes.jpg')
    edges_detected = cv2.Canny(img, 255, 255)
    contours, h = cv2.findContours(edges_detected, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    # for c in contours:
    #     print(c)
    index = 1
    for contour  in contours:
        # #find bounding rectangle
        # (x,y,w,h) = cv2.boundingRect(contour)
        # print(f"x:{x},y:{y},w:{w},h:{h}")
        #draw bounding rectangle
        (x,y,w,h) = cv2.boundingRect(contour)
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    img_contour = img[y: y+h, x: x+w]
    cv2.imshow(f"contour_{index}", img_contour)
    # save the contour
    cv2.imwrite(f"/tmp/contour_{index}.jpg", img_contour)
    index += 1
    cv2.imshow('original', img)
    cv2.imwrite('/tmp/original.jpg', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function3()
```

```python
def function4():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set2/someshapes.jpg')
    edges_detected = cv2.Canny(img, 255, 255)
    contours, h = cv2.findContours(edges_detected, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    index = 1
    for contour in contours:
        # find the bounding rectangle
        (x, y, w, h) = cv2.boundingRect(contour)
        # print(f"x: {x}, y: {y}, w: {w}, h: {h}")
        # draw the bounding rectangle
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 2)
        # crop the contour
        img_contour = img[y: y + h, x: x + w]
        cv2.imshow(f"contour_{index}", img_contour)
        # save the contour
        cv2.imwrite(f"/tmp/contour_{index}.jpg", img_contour)
        index += 1
    cv2.imshow('original', img)
    cv2.imwrite('/tmp/original.jpg',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function4()
```

```python
def function6():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set2/numbers.jpg')
    # detect the edges
    img_edged = cv2.Canny(img, 255, 255)
    # find all the contours
    contours, h = cv2.findContours(img_edged, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    # cv2.drawContours(img, contours, -1, (0, 0, 255), 2)
    # print(f"contours: {len(contours)}")
    for c in contours:
        # find the bounding rectangle
        (x, y, w, h) = cv2.boundingRect(c)
        # highlight the rectangle
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        # highlight the contour
        # cv2.drawContours(img, [c], -1, (0, 0, 255), 2)
    cv2.imshow('original', img)
    cv2.imshow('edged', img_edged)
    cv2.imwrite('/tmp/original.jpg', img)
    cv2.imwrite('/tmp/edged.jpg', img_edged)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function6()
```
