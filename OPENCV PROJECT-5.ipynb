```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
#get video capture
capture = cv2.VideoCapture('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/Megamind.avi')
```

```python
#ret : return value
#frame : the captured img from video source
```

```python
ret, frame1 = capture.read()
ret, frame2 = capture.read()
ret, frame3 = capture.read()
```

```python
cv2.imshow('frame1',frame1)
cv2.imshow('frame2',frame2)
cv2.imshow('frame3',frame3)
cv2.waitKey(0)
cv2.destroyAllWindows()xx
```

```python
# screenshots of frame
from IPython.display import Image, display

listOfImageNames = ['/home/sunbeam/Pictures/frame1.png','/home/sunbeam/Pictures/frame2.png',
                   '/home/sunbeam/Pictures/frame3.png']

for imageName in listOfImageNames:
    display(Image(filename=imageName))
```

```python
while True:
    ret, frame = capture.read()
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('live video',frame)
    cv2.imshow('gray video', gray)

    if cv2.waitKey(5) == ord('x'):
        break

cv2.destroyAllWindows()

#release the capture
capture.release()
```
