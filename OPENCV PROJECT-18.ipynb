```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
def function1():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set2/Hillary.jpg')
    # create the classifier
    classifier = cv2.CascadeClassifier(
        '/home/sunbeam/Documents/gitdata/ML/haarcascades/haarcascade_frontalface_default.xml')
    # find the face
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = classifier.detectMultiScale(gray)
    # print(faces)
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.imshow('original', img)
    cv2.imwrite('/tmp/original.jpg',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function1()
```

```python
def function2():
    capture = cv2.VideoCapture(0)
    classifier_face = cv2.CascadeClassifier('/home/sunbeam/Documents/gitdata/ML/haarcascades/haarcascade_frontalface_default.xml')
    classifier_eye = cv2.CascadeClassifier('/home/sunbeam/Documents/gitdata/ML/haarcascades/haarcascade_eye.xml')
    while True:
        ret, frame = capture.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        # find the faces
        faces = classifier_face.detectMultiScale(gray)
        # highlight the faces
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        # find and highlight the eyes
        eyes = classifier_eye.detectMultiScale(gray)
        for (x, y, w, h) in eyes:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)
        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    capture.release()
    cv2.destroyAllWindows()
function2()
```
