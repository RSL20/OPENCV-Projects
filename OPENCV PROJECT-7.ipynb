```python
def translation():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
    (h,w,d) = img.shape
    img_copy = img.copy()
    #crete the translation matrix
    t = np.float32([[1,0,20],[0,1,20]])
    new_image = cv2.warpAffine(img_copy,t,(w,h))
    cv2.imshow('messi_translation', new_image)
    cv2.imwrite('/tmp/messi_translation.jpg',img_copy)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
translation()
```

```python
from IPython.display import Image, display

listOfImageNames = ['/tmp/messi_translation.jpg']

for imageName in listOfImageNames:
    display(Image(filename=imageName))
```

```python
def rotation():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
    (h, w, d) = img.shape
    img_copy = img.copy()
    center = (w//2,h//2)
    #rotation matrix
    t = cv2.getRotationMatrix2D(center,-45,1)
    new_image = cv2.warpAffine(img,t,(w,h))
    cv2.imshow('messi_rotation', new_image)
    cv2.imwrite('/tmp/messi_rotation.jpg',img_copy)    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
rotation()
```

```python
from IPython.display import Image, display

listOfImageNames = ['/tmp/messi_rotation.jpg']

for imageName in listOfImageNames:
    display(Image(filename=imageName))
```

```python
def resize():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
    (h, w, d) = img.shape
    new_image = cv2.resize(img,(w*2,h*2))
    cv2.imshow('messi_resize', new_image)
    cv2.imwrite('/tmp/messi_resize.jpg',new_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
resize()
```

```python
from IPython.display import Image, display

listOfImageNames = ['/tmp/messi_resize.jpg']

for imageName in listOfImageNames:
    display(Image(filename=imageName))
```

```python
def pyramid():
    #original
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
    larger = cv2.pyrUp(img)
    smaller = cv2.pyrDown(img)
    # smaller of smaller
    smaller_smaller = cv2.pyrDown(smaller)
    cv2.imshow('original', img)
    cv2.imshow('larger',larger)
    cv2.imshow('smaller',smaller)
    cv2.imshow('smaller_smaller', smaller_smaller)
    cv2.imwrite('/tmp/original.jpg',img)
    cv2.imwrite('/tmp/larger.jpg',larger)
    cv2.imwrite('/tmp/smaller.jpg',smaller)
    cv2.imwrite('/tmp/smaller_smaller.jpg',smaller_smaller)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
pyramid()
```

```python
def crop():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
    #(200,60),(275,150)
    face = img[60:150,200:275]
    #(330,285),(400,300)
    ball = img[285:340,330:400]
    cv2.imshow('face',face)
    cv2.imshow('ball',ball)
    cv2.imwrite('/tmp/face.jpg',face)
    cv2.imwrite('/tmp/ball.jpg',ball)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
crop()
```
