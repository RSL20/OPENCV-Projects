```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
#read in 3D array
img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
```

```python
from IPython.display import Image, display

listOfImageNames = ['/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg']

for imageName in listOfImageNames:
    display(Image(filename=imageName))
```

```python
#split img into 3 channels
(b,g,r) = cv2.split(img)
print(b)
print(g)
print(r)
```

```python
#merge channels to create new img
img1 = cv2.merge([b,g,r])
img_r = cv2.merge([b,g,r+100])
img_g = cv2.merge([b,g+100,r])
img_b = cv2.merge([b+100,g,r])
```

```python
#show files
cv2.imshow('orginal image',img1)
cv2.imshow('red image',img_r)
cv2.imshow('green image',img_g)
cv2.imshow('blue image',img_b)

#save files
cv2.imwrite('/tmp/original image.jpg',img1)
cv2.imwrite('/tmp/red image.jpg',img_r)
cv2.imwrite('/tmp/green image.jpg',img_g)
cv2.imwrite('/tmp/blue image.jpg',img_b)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
