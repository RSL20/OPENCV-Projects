```python
def function1():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')

    #copy image
    img_copy = img.copy()

    #draw rectangle
    cv2.rectangle(img_copy,(200,60),(275,150),(0,0,255),2)
    cv2.rectangle(img_copy,(330,285),(400,340),(0,255,255),2)

    #show file
    cv2.imshow('messi_rectangle',img_copy)
    
    #save file
    cv2.imwrite('/tmp/messi_rectangle.jpg',img_copy)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()

function1()
```

```python
def function2():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
    img_copy = img.copy()

    #draw circle
    #highlight face
    cv2.circle(img_copy,(235,105),50,(0,255,255),2)
    #highlight football
    cv2.circle(img_copy,(365,315),27,(0,0,255),2)

    cv2.putText(img_copy,"Messi and his football", (30,30), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1,(0,255,255),2)

    #show file
    cv2.imshow('messi_circle',img_copy)
    
    #save file
    cv2.imwrite('/tmp/messi_circle.jpg',img_copy)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()

function2()
```
