```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
#read in 3D array
img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
```

```python
plt.imshow(img)
```

```python
#convert colorful image to gray scale image
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
#show file
cv2.imshow('messi',img)
cv2.imshow('gray',gray)

#save file
cv2.imwrite('/tmp/messi-color.jpg',img)
cv2.imwrite('/tmp/messi-gray.jpg',gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from IPython.display import Image, display

listOfImageNames = ['/tmp/messi-color.jpg',
                    '/tmp/messi-gray.jpg']

for imageName in listOfImageNames:
    display(Image(filename=imageName))
```
