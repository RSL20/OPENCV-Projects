```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
def function1():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set2/opencv_inv.png')
    # kernel to decide the number of pixels added on the boundaries
    kernel = np.ones((5, 5), dtype='uint8')
    # dilate the image
    img2 = cv2.dilate(img, kernel)
    #erose the image
    img3 = cv2.erode(img,kernel)
    cv2.imshow('original',img)
    cv2.imshow('dilate',img2)
    cv2.imshow('erode',img3)
    cv2.imwrite('/tmp/original.jpg', img)
    cv2.imwrite('/tmp/dilate.jpg', img2)
    cv2.imwrite('/tmp/erode.jpg', img3)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function1()
```

```python
def function3():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set2/Origin_of_Species.jpg')
    # gray
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # threshold
    ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
    # dilate the image
    kernel = np.ones((2,2), dtype='uint8')
    dilated = cv2.dilate(thresh,kernel)
    cv2.imshow('original', img)
    cv2.imshow('thresh', thresh)
    cv2.imshow('dilated', dilated)
    cv2.imwrite('/tmp/original.jpg', img)
    cv2.imwrite('/tmp/thresh.jpg', thresh)
    cv2.imwrite('/tmp/dilated.jpg', dilated)    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function3()
```
