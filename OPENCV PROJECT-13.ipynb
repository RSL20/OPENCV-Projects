```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
def function1():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set1/gradient.png')
    # threshold the image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, img2 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    ret, img3 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
    cv2.imshow('original', img)
    cv2.imshow('threshold', img2)
    cv2.imshow('threshold2', img3)
    cv2.imwrite('/tmp/original.jpg', img)
    cv2.imwrite('/tmp/threshold.jpg', img2)
    cv2.imwrite('/tmp/threshold2.jpg', img3)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function1()
```

```python
from IPython.display import Image, display

listOfImageNames = ['/tmp/original.jpg','/tmp/threshold.jpg','/tmp/threshold2.jpg']

for imageName in listOfImageNames:
    display(Image(filename=imageName))
```

```python
def function2():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set2/Origin_of_Species.jpg')
    #gray
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    #threshold
    ret, img_thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
    ret, img_thresh2 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
    cv2.imshow('original',img)
    cv2.imshow('threshold',img_thresh)
    cv2.imshow('threshold2',img_thresh2)
    cv2.imwrite('/tmp/original.jpg', img)
    cv2.imwrite('/tmp/threshold.jpg', img_thresh)
    cv2.imwrite('/tmp/threshold2.jpg', img_thresh2)    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function2()
```
