```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
def function1():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set2/scan.jpg')
    points_A = np.float32([[320, 15], [700, 215], [85, 610], [530, 780]])
    points_B = np.float32([[0, 0], [420, 0], [0, 594], [420, 594]])
    M = cv2.getPerspectiveTransform(points_A, points_B)
    warped = cv2.warpPerspective(img, M, (420, 594))
    cv2.imshow('original', img)
    cv2.imshow('warped', warped)
    cv2.imwrite('/tmp/original.jpg', img)
    cv2.imwrite('/tmp/warped.jpg', warped)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function1()
```

```python
from IPython.display import Image, display

listOfImageNames = ['/tmp/original.jpg','/tmp/warped.jpg']

for imageName in listOfImageNames:
    display(Image(filename=imageName))
```
