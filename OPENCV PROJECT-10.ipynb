```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
def function1():
    img1 = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
    img2 = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/opencv-logo-white.png')
    img1_resized = cv2.resize(img1,(512,512))
    img2_resized = cv2.resize(img2,(512,512))
    # new_image = cv2.add(img1_resized, img2_resized)
    new_image = cv2.addWeighted(img1_resized,0.7,img2_resized,0.3,0.5)
    cv2.imshow('new image', new_image)
    cv2.imwrite('/tmp/new image.jpg', new_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function1()
```

```python
def function2():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/day10_13dec/images/messi5.jpg')
    m = np.ones(img.shape, dtype='uint8') * 100
    new_image_add = cv2.add(img, m)
    new_image_subtract = cv2.subtract(img, m)
    cv2.imshow('new image - add', new_image_add)
    cv2.imshow('new image - subtract', new_image_subtract)
    cv2.imwrite('/tmp/new image - add.jpg', new_image_add)
    cv2.imwrite('/tmp/new image - subtract.jpg', new_image_subtract)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function2()
```
