```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
def function1():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set1/messi5.jpg')
    # create the kernel matrix
    # please normalize the values
    kernel = np.ones((10, 10), dtype='float32') / 100
    # perform the blurr the operation
    blurred = cv2.filter2D(img, -1, kernel)
    cv2.imshow('original', img)
    cv2.imshow('blurred', blurred)
    cv2.imwrite('/tmp/messi_blurred.jpg', blurred)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function1()
```

```python
def function2():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set1/messi5.jpg')
    # create the kernel matrix
    # please normalize the values
    kernel = np.ones((7, 7), dtype='float32') / 49
    # extract the face
    face = img[60:150, 200:275]
    # perform the blurr the operation
    blurred = cv2.filter2D(face, -1, kernel)
    cv2.imshow('original', img)
    cv2.imshow('blurred', blurred)
    cv2.imwrite('/tmp/messi_blurred.jpg', blurred)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function2()
```

```python
def function3():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set1/messi5.jpg')
    #blurr image
    #blurred = cv2.blur(img,(5,5))
    blurred = cv2.GaussianBlur(img, (9, 9), 0)
    cv2.imshow('original', img)
    cv2.imshow('blurred', blurred)
    cv2.imwrite('/tmp/messi_blurred.jpg', blurred)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function3()
```

```python
def function4():
    img = cv2.imread('/home/sunbeam/Documents/gitdata/ML/dataset/images2/set1/messi5.jpg')
    # kernel for sharpening the image
    kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
    # get the sharpened image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sharpened2 = cv2.filter2D(gray, -1, kernel)
    sharpened = cv2.filter2D(img, -1, kernel)
    cv2.imshow('original', img)
    cv2.imshow('sharpened', sharpened)
    cv2.imshow('gray_sharpened', sharpened2)
    cv2.imwrite('/tmp/original.jpg', img)
    cv2.imwrite('/tmp/sharpened.jpg', sharpened)
    cv2.imwrite('/tmp/gray_sharpened.jpg', sharpened2)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
function4()
```
